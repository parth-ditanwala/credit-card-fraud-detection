```python
# importing required modules and libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# removing outliers
df.drop(df[df["V5"]<-50].index,axis = 0,inplace = True)
df.drop(df[df["V6"]> 35].index,axis = 0,inplace = True)
df.drop(df[df["V7"]> 50].index,axis = 0,inplace = True)
```

```python
# spliting dataframe into target variable and features
x = df.drop(columns = ["Class"])
y = df["Class"]
```

```python
# balancing the proportions of catagories using SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy = 0.5)
x_smote, y_smote = smote.fit_resample(x,y)
y_smote.value_counts()
```

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.model_selection import RepeatedStratifiedKFold

classifier = RandomForestClassifier(n_estimators = 20, max_depth = 50)
splits = RepeatedStratifiedKFold().split(x_smote,y_smote)
validation = cross_validate(classifier,x_smote,y_smote,scoring =["accuracy","f1","roc_auc"], cv =splits)
validation
```

```python
score_df = pd.DataFrame(validation)
score_df
```

```python
# since we have validated our results using validation sets we can build a final model now
#splitting the dataset into train and test sets

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
x_train,x_test,y_train,y_test = train_test_split(x_smote,y_smote,test_size = 0.2, stratify = y_smote,random_state = 10)

#fitting a model
classifier = RandomForestClassifier(n_estimators = 20,max_depth = 50)
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)

print("model accuracy :{}".format(classifier.score(x_test,y_test)))
print("F1_score : {}".format(f1_score(y_test,y_pred)))
```
